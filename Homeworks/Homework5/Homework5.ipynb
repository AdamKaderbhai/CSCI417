## Problem 4

Given two matrices A, B of size m Ã—n and n Ã—p respectively (implemented
using numpy arrays). Write a function named matrix multiplication(A,B) that returns
that product AB. For this problem, please do not use the np.dot() method. The goal of
this exercise is to understand how matrix multiplication works.

```python
def matrix_multiplication(A, B):
    # Dimensions of matrices A and B
    m = len(A)
    n = len(A[0])
    p = len(B[0])

    # Initialize the result matrix with zeros
    result = [[0 for _ in range(p)] for _ in range(m)]

    # Perform the multiplication
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]

    return result
    
# Test with example matrices

A = [[1, 2], 
     [3, 4]]

B = [[5, 6], 
     [7, 8]]
print(matrix_multiplication(A, B))
```

## Problem 6

The following problem is a modification of the lattice problem that we did
in class.
You are on a 2D grid, starting at the origin point (0, 0). Your objective is to reach the
point (m, n) using specific movement rules. You can move in the following ways:
â€¢ Rightward Move: Move one step right from (x, y) to (x + 1, y).
â€¢ Upward Move: Move one step up from (x, y) to (x, y + 1).
â€¢ Diagonal Move: Move one step diagonally up and to the right from (x, y) to
(x + 1, y + 1).
Let A(m, n) represent the total number of distinct paths you can take to reach the point
(m, n) from the origin (0, 0) using any combination of the allowed moves. Find A(m, n).
What is the big-O performance of your algorithm?
For example, for (m, n) = 1 we have A(1, 1) = 3. All distinct paths are
(1) Path 1: (0, 0) â†’(1, 1) (Diagonal Move)
(2) Path 2: (0, 0) â†’(1, 0) â†’(1, 1) (Right then Up)
(3) Path 3: (0, 0) â†’(0, 1) â†’(1, 1) (Up then Right)
3


```python
def count_paths(m, n):
    # Base case:
    if m == 0 or n == 0:
        return 1

    # Recursive case:
    return count_paths(m - 1, n) + count_paths(m, n - 1) + count_paths(m - 1, n - 1)

# Example
m, n = 1, 1
result = count_paths(m, n)
print(f"A({m}, {n}) =", result)
```
